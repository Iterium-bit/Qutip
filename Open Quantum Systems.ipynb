Problem: Damped Rabi Oscillations

Physics:
We have a qubit driven by a field with Hamiltonian H = sigma_x.
At the same time, the qubit loses energy to the environment through a decay process C = sqrt(gamma) * sigma_-.
The driving field tries to push the qubit toward the excited state |1>, while the decay tries to bring it back to the ground state |0>.
